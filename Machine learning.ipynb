```python
import pandas as pd
```

```python
cartier = pd.read_csv('dataset/cartier_catalog.csv')
```

```python
cartier.drop(['ref' , 'image' , 'description'] , axis  = 1 , inplace=True)
```

```python
cartier.head(5)
```

```python
cartier["tags"] = cartier["tags"].str.replace(".","").str.replace("-", "_")
```

```python
def tag_spiliter(dataframe , col_name , delimiter , metal , first_gem , second_gem , third_gem , fourth_gem):
    dataframe['str_split'] = dataframe[col_name].str.split(delimiter)
    dataframe[metal] = dataframe.str_split.str.get(0).str.strip()
    dataframe[first_gem] = dataframe.str_split.str.get(1).str.strip()
    dataframe[second_gem] = dataframe.str_split.str.get(2).str.strip()
    dataframe[third_gem] = dataframe.str_split.str.get(3).str.strip()
    dataframe[fourth_gem] = dataframe.str_split.str.get(4).str.strip()
    dataframe.fillna(0 , inplace=True)
    del dataframe['str_split']
```

```python
tag_spiliter(cartier , 'tags' , ',' , 'metal' , 'first_gem' , 'second_gem' , 'third_gem' , 'fourth_gem')
```

```python
def dummies_concat(col_list, dataframe):
    list_of_dummies = []
    for material in col_list:
        list_of_dummies.append(cartier[material].str.get_dummies())
    dummies_df = pd.concat(list_of_dummies , sort=True).groupby(level=0).any().astype(int)
    dataframe = dataframe.join(dummies_df)
    return dataframe
```

```python
list_of_material = ['categorie', 'metal' , 'first_gem' , 'second_gem' , 'third_gem' , 'fourth_gem']
```

```python
ready_df_for_ML = dummies_concat(list_of_material , cartier)
```

```python
ready_df_for_ML.drop(['tags' , 'metal' ,'first_gem' ,'second_gem' ,'third_gem', 'fourth_gem' , '0' , 'title' ,'categorie'] , axis = 1 , inplace=True)
```

```python
ready_df_for_ML.head(10)
```

```python
ready_df_for_ML.to_csv('ready_for_ml.csv' , index = False)
```
